## 4.3.- Funciones con M煤ltiples Par谩metros

Ahora armaremos una funci贸n usando IF que nos dice cual es la raiz cuadrada m谩s grande de 3 n煤meros

```python
def raiz_max(num1, num2, num3):
    """calculemos la raiz mas grande de 3 numeros"""
    res_num1 = num1**(1/2)
    res_num2 = num2**(1/2)
    res_num3 = num3**(1/2)
    raiz_max = res_num1
    if res_num2 > raiz_max:
        raiz_max = res_num2
    if res_num3 > raiz_max:
        raiz_max = res_num3
    return raiz_max
numero1 = int(input("dame un numero1"))
numero2 = int(input("dame un numero2"))
numero3 = int(input("dame un numero3"))
raiz_max(numero1, numero2, numero3)
```

Por si las dudas, esto se podr铆a haber simplificado usando Max y Min

```python
def raiz_max(num1, num2, num3):
    """calculemos la raiz mas grande de 3 numeros"""
    return max(num1**(1/2), num2**(1/2), num3**(1/2))  # Usamos el max :U
numero1 = int(input("dame un numero1"))
numero2 = int(input("dame un numero2"))
numero3 = int(input("dame un numero3"))
raiz_max(numero1, numero2, numero3)
```

Calcula las medidas de tendencia central

```python
import statistics as estad
import random

def calificaiones1():
 lista_calif = [] 
 for calif in range(30):
  calificaciones = random.randrange(50, 100)
  lista_calif.append(calificaciones)
 return lista_calif
 
valor_calif = calificaiones1()
print(valor_calif)

# Medidas de tendencia central
media = estad.mean(valor_calif)
mediana = estad.median(valor_calif)
moda = estad.mode(valor_calif)

print('Las medidas de tendencia central son:')
print(f'Media= {media:>7}')
print(f'Mediana= {mediana:>5}')
print(f'Moda= {moda:>6}')
```

Calcula las medidas de dispersi贸n

```python
import statistics as estad
import random

def calificaiones1():
 lista_calif = [] 
 for calif in range(30):
  calificaciones = random.randrange(50, 100)
  lista_calif.append(calificaciones)
 return lista_calif
 
valor_calif = calificaiones1()
print(valor_calif)

# Medidas de dispersi贸n
rango=max(valor_calif)-min(valor_calif)
varianza=estad.pvariance(valor_calif)
desv_estandar=estad.pstdev(valor_calif)

print('Las medidas de dispersi贸n son:')
print(f'Rango= {rango:>2}')
print(f'Varianza= {varianza:>0.3f}')
print(f'Desv. Est谩ndar= {desv_estandar:>0.3f}')
```
